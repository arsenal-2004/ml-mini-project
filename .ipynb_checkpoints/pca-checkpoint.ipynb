```python
import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Create Data Tensor
X, y, class_names = preprocessing.create_data_tensor()
```

```python
#Convert from 3D to 2D for PCA
X_flat = preprocessing.flatten_data(X)
```

```python
#Scale data for better representation
sc = StandardScaler()
X= sc.fit_transform(X_flat)
```

```python
#Applying PCA
pca = PCA(n_components= 3) #To represent data in 3D
X = pca.fit_transform(X_flat)
```

```python
X.shape
```

```python
#Initialize empty 3D graph
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
#Plot points for first 10 classes
c=['yellow', 'green', 'blue', 'pink', 'red', 'cyan', 'brown', 'orange', 'grey', 'magenta']
for classes in range(0,10):
    for i in range(classes*27,(classes+1)*27+1):
        ax.scatter(xs=X[i,0],ys=X[i,1], zs=X[i,2], c=c[classes], s=10)
```
